```python
import sklearn
import numpy as np
import pandas as pd
import plotly.express as px

n=1000
n_features=4
variance=1


x1 = np.random.multivariate_normal([0, 0, 0],
 variance*np.array([[10, 3, 1],
  [3, 5, -1],
  [1, -1, 1]]), n//2)


x2 = np.random.multivariate_normal([1, 1/2, 2],
 variance*np.array([[10, 3, 1],
  [3, 5, -1],
  [1, -1, 1]]), n//2)


x = np.concatenate([x1, x2])
x = np.concatenate([np.ones([n, 1]), x], axis=1)

y = np.concatenate([np.ones(n//2, dtype='int'), np.zeros(n//2, dtype='int')])

x, y = sklearn.utils.shuffle(x, y)

df = pd.DataFrame(np.concatenate([y.reshape((n,1)), x], axis=1), columns=["y", "x0", "x1", "x2", "x3"])

fig = px.scatter_3d(df, x='x1', y='x2', z='x3', color='y', size="x0", size_max=10)
fig.show()

wei=np.random.random(n_features)
```

```python
def activation(x):
  return np.array([1 if i > 0.5 else 0 for i in x])

lr=0.1

for e in range(100):
  y_pred = activation(x.dot(wei))
  wei += -lr*x.T.dot(y_pred-y)

#accuracy
(y_pred==y).mean()
```
